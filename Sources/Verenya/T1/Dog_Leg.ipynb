```python
import sys
import numpy as np
sys.path.append('../../')
import helper
```

```python
dataset_name = '2'
dataset_params = helper.get_params_dataset(dataset_name, 'Research/')
dataset_filename = helper.get_filenames_datasets('Research/')[dataset_name]
f = helper.get_func_research(dataset_params['f_label'])

print(dataset_params)
print(dataset_filename)

X, Y, datasets = helper.load_datasets(dataset_filename)

init_weights = np.array([20, 3], dtype=float)
```

```python
result_weights, count_step = helper.dog_leg(f, datasets[0][:, 0], datasets[0][:, 1], init_weights, epsilon=1e-12, max_iter=100, radius=1.0)
result_loss = helper.mse_loss(datasets[0][:, 0], datasets[0][:, 1], result_weights, f)
print(result_loss)
print(result_weights, count_step)
helper.print_generated(X, Y, datasets[0], f(X, result_weights))
```

```python
result_weights, count_step = helper.gauss_newton_fast(f, datasets[0][:, 0], datasets[0][:, 1], init_weights, epsilon=2e-2, max_iter=100)
result_loss = helper.mse_loss(datasets[0][:, 0], datasets[0][:, 1], result_weights, f)
print(result_loss)
print(result_weights, count_step)   
helper.print_generated(X, Y, datasets[0], f(X, result_weights))
```
