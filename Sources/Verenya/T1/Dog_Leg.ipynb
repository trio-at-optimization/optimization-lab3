```python
def f(x, w):
    return np.sin(w[0] * x) + w[1]

datadasets_filename = helper.get_filenames_datasets('Research/')
print(datadasets_filename)
X, Y, datasets = helper.load_datasets(datadasets_filename['1'])

init_weights = np.array([3, 4], dtype=float)
```

```python
result_weights, count_step = helper.dog_leg(f, datasets[0][:, 0], datasets[1][:, 0], init_weights, epsilon=2e-2, max_iter=100)
print(result_weights, count_step)
helper.print_generated(X, Y, datasets[0], f(X, result_weights))
```
